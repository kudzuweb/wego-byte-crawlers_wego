```python
#import packages

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in csv

wego = pd.read_csv("../data/headway_data_clean.csv")
```

```python
#change dates to datetime64 objects

wego['DATE'] = pd.to_datetime(wego['DATE'],
                format = "%Y/%m/%d")
```

```python
#create 'DAYS_OF_THE_WEEK' column
wego['DAYS_OF_THE_WEEK'] = wego['DATE'].dt.weekday
wego.head()
```

```python
# create 'MONTH' column

wego['MONTH'] = pd.DatetimeIndex(wego['DATE']).month
wego
```
