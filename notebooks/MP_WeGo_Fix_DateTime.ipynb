```python
#import packages
import pandas as pd
pd.set_option("display.max_columns", 100)
import numpy as np
import seaborn as sns
```

```python
#read in csv & display
wego = pd.read_csv('../data/headway_data.csv')
wego
```

```python
#how many times does the messed up time appear? a lot

# wego['SCHEDULED_TIME'].value_counts()['1900-01-01 00:08:00']
```

```python
wego['NextDay_Scheduled'] = np.where(wego['SCHEDULED_TIME'].astype(str).str.contains('1900'), '1', '0')
wego['NextDay_Actual_Arrival'] = np.where(wego['ACTUAL_ARRIVAL_TIME'].astype(str).str.contains('1900'), '1', '0')
wego['NextDay_Actual_Departure'] = np.where(wego['ACTUAL_DEPARTURE_TIME'].astype(str).str.contains('1900'), '1', '0')

wego
```

```python
wego.loc[wego['NextDay_Scheduled']=='1', 'SCHEDULED_TIME']
```

```python
#replace broken times with NaN
wego['SCHEDULED_TIME'] = wego['SCHEDULED_TIME'].str.replace('1900-01-01 ','')
wego['ACTUAL_ARRIVAL_TIME'] = wego['ACTUAL_ARRIVAL_TIME'].str.replace('1900-01-01 ','')
wego['ACTUAL_DEPARTURE_TIME'] = wego['ACTUAL_DEPARTURE_TIME'].str.replace('1900-01-01 ','')
```

```python
# concatenate dates onto each time column to prevent '1900' issue

wego['SCHEDULED_TIME'] = wego['DATE'] + ' ' + wego['SCHEDULED_TIME']
wego['ACTUAL_ARRIVAL_TIME'] = wego['DATE'] + ' ' + wego['ACTUAL_ARRIVAL_TIME']
wego['ACTUAL_DEPARTURE_TIME'] = wego['DATE'] + ' ' + wego['ACTUAL_DEPARTURE_TIME']
wego
```

```python
#convert dates to datetime64 objects, check dtype to confirm

wego['DATE'] = pd.to_datetime(wego['DATE'], format = "%Y-%m-%d")
wego['DATE'].info()
```

```python
# convert times to datetime64 objects, check dtype to confirm
wego['SCHEDULED_TIME'] = pd.to_datetime(wego['SCHEDULED_TIME'], format = '%Y-%m-%d %H:%M:%S')
wego['ACTUAL_ARRIVAL_TIME'] = pd.to_datetime(wego['ACTUAL_ARRIVAL_TIME'], format = '%Y-%m-%d %H:%M:%S')
wego['ACTUAL_DEPARTURE_TIME'] = pd.to_datetime(wego['ACTUAL_DEPARTURE_TIME'], format = '%Y-%m-%d %H:%M:%S')
wego[['SCHEDULED_TIME', 'ACTUAL_ARRIVAL_TIME', 'ACTUAL_DEPARTURE_TIME']].info()
```

```python
wego.loc[wego['NextDay_Scheduled']=='1', 'SCHEDULED_TIME'] = wego.loc[wego['NextDay_Scheduled']=='1', 'SCHEDULED_TIME'] + pd.Timedelta('1 day')
wego.loc[wego['NextDay_Actual_Arrival']=='1', 'ACTUAL_ARRIVAL_TIME'] = wego.loc[wego['NextDay_Actual_Arrival']=='1', 'ACTUAL_ARRIVAL_TIME'] + pd.Timedelta('1 day')
wego.loc[wego['NextDay_Actual_Departure']=='1', 'ACTUAL_DEPARTURE_TIME'] = wego.loc[wego['NextDay_Actual_Departure']=='1', 'ACTUAL_DEPARTURE_TIME'] + pd.Timedelta('1 day')
```

```python
#export to csv

wego.to_csv('../data/headway_data_clean.csv', index=False)
```
